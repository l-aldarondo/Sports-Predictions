### ETL_wnba_stats
### loading data to PostgreSQl database wnba_champs

```python
# Import our dependencies
import pandas as pd
from path import Path
from sqlalchemy import create_engine
from config import db_password
```

```python
# Load the season_stats.csv dataset.
csv_path = Path("./Data/wmba_stats.csv")
stats_df = pd.read_csv(csv_path)
stats_df.head()
```

```python
# remove usless columns
stats_df = stats_df.drop(['Team'], axis=1)
stats_df.head()
```

```python
# renaming columns
stats_df = stats_df.rename(columns={'W/L%': 'Win_pct', 'eFG%': 'eFG_pct','TS%': 'TS_pct', 'TOV%': 'TOV_pct', 'ORB%': 'ORB_pct' ,'FT/FGA': 'FT_FGA',
                                   'opp_eFG%': 'Opp_eFG_pct','opp_TS%': 'Opp_TS_pct', 'opp_TOV%': 'Opp_TOV_pct', 'opp_ORB%': 'Opp_ORB_pct', 'opp_FT/FGA': 'Opp_FT_FGA' })
stats_df.head()
```

```python
#Load the cy_stats.csv dataset (current year).
csv_path = Path("./Data/cy_wnba_stats.csv")
cy_stats_df = pd.read_csv(csv_path)
cy_stats_df
```

```python
# remove usless columns
cy_stats_df = cy_stats_df.drop(['Team'], axis=1)
cy_stats_df.head()
```

```python
# renaming cy_stats.cvs (current year).
cy_stats_df = cy_stats_df.rename(columns={'W/L%': 'Win_pct', 'eFG%': 'eFG_pct','TS%': 'TS_pct', 'TOV%': 'TOV_pct', 'ORB%': 'ORB_pct' ,'FT/FGA': 'FT_FGA',
                                   'opp_eFG%': 'Opp_eFG_pct','opp_TS%': 'Opp_TS_pct', 'opp_TOV%': 'Opp_TOV_pct', 'opp_ORB%': 'Opp_ORB_pct', 'opp_FT/FGA': 'Opp_FT_FGA' })
cy_stats_df.head()
```

```python
# Load the cy_stats2.csv dataset.(dummy year)
csv_path = Path("./Data/cy_wnba_stats2.csv")
cy_stats2_df = pd.read_csv(csv_path)
cy_stats2_df
```

```python
# renaming columns cy_stats2
cy_stats2_df = cy_stats2_df.rename(columns={'W/L%': 'Win_pct', 'eFG%': 'eFG_pct','TS%': 'TS_pct', 'TOV%': 'TOV_pct', 'ORB%': 'ORB_pct' ,'FT/FGA': 'FT_FGA',
                                   'opp_eFG%': 'Opp_eFG_pct','opp_TS%': 'Opp_TS_pct', 'opp_TOV%': 'Opp_TOV_pct', 'opp_ORB%': 'Opp_ORB_pct', 'opp_FT/FGA': 'Opp_FT_FGA' })
cy_stats2_df.head()
```

```python
# remove usless columns
cy_stats2_df = cy_stats2_df.drop(['Team'], axis=1)
cy_stats2_df.head()
```

### Initializing database connection

```python
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/wnba_champs"
```

```python
engine = create_engine(db_string)
```

### Loading data to wnba_champs database

```python
stats_df.to_sql(name='stats', con=engine)
```

```python
cy_stats_df.to_sql(name='cy_stats', con=engine)
```

```python
cy_stats2_df.to_sql(name='cy_stats2', con=engine)
```
